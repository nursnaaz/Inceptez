```python
import sys
a = [1,2,3,40,0,23]
b = [1,3,4,0,2,4]

for x in a:
    for y in b:
        try:
            print(x)
            print(y)
            print("\n")
            x/y
        except:
            print("OOPs!",sys.exc_info()[0],"occured" )
            print()
        finally:
            print("Try is done and it is executed finally")
            
```
