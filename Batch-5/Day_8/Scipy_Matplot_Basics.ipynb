```python
#Scipy

from scipy.misc import imread, imresize, imsave,imshow

img = imread('test.jpg')
print(img.dtype)
print(img.shape)

img_tinted = img *(1.0 , 0.90, 0.95)

print(img_tinted.dtype)

img_tinted = imresize(img_tinted,(300,300))

imsave('test_resize.jpg', img_tinted)


```

```python
#MatplotLib
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,3*np.pi,0.1)

print(x)

y = np.cos(x)
print(y)

plt.plot(x,y)

plt.show()

y = np.sin(x)
print(y)

plt.plot(x,y)

plt.show()

y = np.tanh(x)
print(y)

plt.plot(x,y)

plt.show()

plt.imshow(np.uint8(img_tinted))
plt.show()

plt.imshow(np.uint8(img))
plt.show()
```
